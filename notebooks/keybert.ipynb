```python
import sys
import os
sys.path.append(os.path.abspath(".."))
from keybert import KeyBERT
from elastic import ScrollingCorpus, elasticsearch_client
from itertools import islice
from pprint import pprint
```

```python
def batched(iterable, n, *, strict=False):
    # batched('ABCDEFG', 3) â†’ ABC DEF G
    if n < 1:
        raise ValueError('n must be at least one')
    iterator = iter(iterable)
    while batch := tuple(islice(iterator, n)):
        if strict and len(batch) != n:
            raise ValueError('batched(): incomplete batch')
        yield batch
```

```python
client = elasticsearch_client("../credentials.json", "../http_ca.crt")
batch_size = 100

corpus = ScrollingCorpus(client, "arxiv-index", batch_size=batch_size, doc_field="summary")
batched_corpus = batched(corpus, batch_size)
model = KeyBERT()

keywords = []

for batch in batched_corpus:
    keywords += model.extract_keywords(batch, use_maxsum=True, use_mmr=True, keyphrase_ngram_range=(1,2))
```

```python
for k in keywords:
    print(k)
```
